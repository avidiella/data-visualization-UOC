# Preprocessament de les dades

## Càrrega de les dades

```python
import pandas as pd

file_name =  "Radiancia_SupArtif_2012_v3_sinvinculos_ASM_zenodo.xls"
sheet =  "ExtraidoRanking"

df_ranking = pd.read_excel(io=file_name, sheet_name=sheet)

df_ranking.head(2)
```

## Uniformització del nom dels municipis

```python
import re

df_ranking["Nom_municipi"] = ""

for idx, i in enumerate(df_ranking["Nombre"]):
    if re.search("\ \((L\'|El|A|La|L(e|o|a)s)\)", i):
        det1 = re.search("\ \((L\'|El|A|La|L(e|o|a)s)\)", i).group(1)
        det = re.search("\ \((L\'|El|A|La|L(e|o|a)s)\)", i).group(0)
        remove_det = i.replace(det, "")
        name_list = [det1, remove_det]
        if det1 != None:
            if det1 == "L'":
                new_name = "".join(name_list)
            else:
                new_name = " ".join(name_list)
    else:
        new_name = i
    df_ranking.iloc[idx,14] = new_name

df_ranking.head(2)
```

Un cop realitzades les modificacions manualment, tornem a carregar les dades per a realitzar els càlculs.

```python
df_calc = pd.read_excel(io="data_cleaning.xls", sheet_name="Sheet1", index_col=0)
```

## Càlculs

### Potència per habitant (2015)

```python
df_calc["Potencia_habitant_2015"] = df_calc["Potencia_emesa"] / df_calc["Poblacio_2015"]
```

### Potència per habitant (2023)

```python
df_calc["Potencia_habitant_2023"] = df_calc["Potencia_emesa"] / df_calc["Poblacio_2023"]
```

### Potència per punt de llum

```python
df_calc["Potencia_PL_2015"] = df_calc["Potencia_emesa"] / df_calc["Punts_llum_estimats"]
```

### Punts de llum per habitant (2015)

```python
df_calc["PL_per_habitant_2015"] = df_calc["Punts_llum_estimats"] / df_calc["Poblacio_2015"]
```

### Estimació de punts de llum estimats per habitant (2023)

```python
df_calc["PL_estimats_habitant_2023"] = df_calc["Punts_llum_estimats"] / df_calc["Poblacio_2023"]
```

### Punts de llum per superfície (2015)

```python
df_calc["PL_superficie"] = df_calc["Punts_llum_estimats"] / df_calc["Superficie_total"]
```

### Punts de llum per superfície artificial

```python
df_calc["PL_superficie_art"] = df_calc["Punts_llum_estimats"] / df_calc["Superficie_artificial"]
```
